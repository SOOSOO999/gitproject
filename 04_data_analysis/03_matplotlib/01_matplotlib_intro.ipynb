### matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### 기본 사용법

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y)
plt.show()
```

##### Figure | Axes

- Figure : 전체 그림이 그려지는 컨테이너
- Axes : 그래프가 실제로 그려지는 영역 (좌표, 데이터)

```python
fig = plt.figure(figsize=(6, 4), facecolor='green')
print(fig)
ax = plt.gca()  # axes의 객체를 가져옴
ax.plot([1, 2, 3],[4, 5, 6])
print(ax)

plt.title('Hello Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 6)) # nrows : 행 수, ncols = 열 수, size 1000 x 600 / 기본적으로 fig, ax 값을 반환
print(fig)
print(ax)
plt.show()
```

##### plt 속성 설정

```python
# label : 기본적으로 한글 폰트 지원하지 않음
plt.plot(x, y)
plt.title('제목')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```

```python
# ticks (눈금)
plt.figure(figsize=(3, 4))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.xticks(ticks=np.arange(0, 101, 5), rotation=90)
plt.yticks(rotation=45)

plt.plot(x, y)
plt.show()
```

```python
# lim (보여지는 범위 재지정)
plt.figure(figsize=(3, 2))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.xlim(1, 50)
plt.ylim(2, 100)

plt.plot(x, y)
plt.show()
```

```python
# legend(범례)
plt.figure(figsize=(3, 2))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)
y2 = np.linspace(3, 300, 100)

plt.plot(x, y, label='시간')
plt.plot(x, y2, label='공간')

# loc 범례 위치 (기본 값: best / 알아서 최적의 위치를 찾아서 위치한다)
# - (upper | lower) (left | right)
# - center

plt.legend(loc='upper left')    

plt.show()
```

##### 그래프 여러 개 그리기기

```python
plt.figure(figsize=(3, 2))

x = np.arange(1, 11)
y = x * 3
y2 = x * 2

plt.plot(x, y, label='선', color='red', marker='o')
plt.bar(x, y2, label='바', color='green')

plt.legend()
plt.grid(True)  # 격자. 기본값은 False

plt.show()
```

```python
# subplots 이용 -> 여러 그래프를 각각 그리기

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x1 = np.arange(1, 11)
y1 = x1 * 2
x2 = np.arange(11, 20)
y2 = x2 * 2

ax[0].plot(x1, y1, label='선', color='red')
ax[0].set_xlabel('X1')  # .set이 객체지향적인 방법
ax[0].set_ylabel('Y1')
ax[0].set_title('빨간 선')
ax[0].legend()
ax[0].grid()
ax[1].bar(x2, y2, label='바', color='yellowgreen')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
ax[1].set_title('연두 막대')
ax[1].legend()
ax[1].grid()
plt.show()
```
