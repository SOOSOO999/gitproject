```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
font_path = 'C:\\Windows\\Fonts\\gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()    # 폰트의 경로로부터 폰트를 가져오고 get_name으로 이름을 가져온다.
matplotlib.rc('font', family=font)  # family= font의 이름을 가진 폰트로 변경
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
iris_data.keys()
# iris_data.target        # 정답. 0, 1, 2 로 이루어진 배열
# iris_data.target_names  
iris_data.feature_names
```

```python
print(iris_data.data)   # input data (특성 데이터)
print(iris_data.feature_names)  # 특성명
```

```python
print(iris_data.target)         # 숫자로 인코딩된 label 데이터 (정답 데이터)
print(iris_data.target_names)   # label명
```

```python
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df['label'] = iris_data.target
iris_df.describe()
```

```python
# 데이터 전처리

# 1. 입력-라벨 데이터
input_data = iris_data.data
input_label = iris_data.target 

# 2. 훈련-테스트 데이터
train_input, test_input, train_label, test_label = train_test_split(input_data, input_label, test_size=0.3, stratify=input_label, random_state=10)   # 

# 3. 정규화
iris_scaler = StandardScaler()
train_scaled = iris_scaler.fit_transform(train_input)
test_scaled = iris_scaler.transform(test_input)
```

```python
# 훈련
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)
```

```python
# 평가
kn.score(test_scaled, test_label)
```

```python
# 평가 : accuracy_score
from sklearn.metrics import accuracy_score
pred = kn.predict(test_scaled)
accuracy_score(test_label, pred)
```

```python
# 예측
predict_input = [5.1, 2.5, 3.6, 1.8]
predict_sclaed = iris_scaler.transform([predict_input])
kn.predict(predict_sclaed)
```
